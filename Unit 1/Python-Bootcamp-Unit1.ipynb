### Installing Python

Before we can get started working with Python, we need to install it. Fortunately, Python is free, open-source software (FOSS).

There are currently two major versions of Python in use: Python 2 and Python 3. However, given Python 2 is [end-of-life (EOL)](https://www.python.org/dev/peps/pep-0373/), Python 3 is highly recommended at this point (side note: for Python 2 to 3 code conversion, consider [2to3](https://docs.python.org/2/library/2to3.html)).

There are many ways to install Python 3, and options vary by your operating system (OS). One simple way to get the latest version is to visit the official Python [website](https://www.python.org). From there, you can download a version for [Mac](https://www.python.org/downloads/mac-osx/), [Windows](https://www.python.org/downloads/windows/), or [Linux](https://www.python.org/downloads/source/).

Alternatively, you can use a package manager to install Python and keep it up-to-date; for example:

- Mac: [Homebrew](https://brew.sh/)
- Windows: [Chocolatey](https://chocolatey.org/); [winget](https://winget.run/pkg/Python/Python);  [Microsoft Store](https://www.microsoft.com/en-us/p/python-39/9p7qfqmjrfp7#activetab=pivot:overviewtab)
- Linux: apt (Ubuntu); Arch (pacman); etc.

Other options include a distribution such as:

- [Anaconda](https://www.anaconda.com/distribution/)
- [Miniconda](https://docs.conda.io/projects/conda/en/latest/user-guide/install/)

Or you can use a virtual environment tool such as [pyenv](https://github.com/pyenv/pyenv), which will let you easily manage multiple Python installations on your computer.

Ultimately, how you get Python installed is up to you!

#### Running Python

While the focus of this tutorial will be on how to work with Python via a JupyterLab, it's worth noting you could run Python directly from a shell environment. The specific shell you would use depends on your computer's operating system (OS); typical options include:

- Mac: Terminal (e.g., Terminal.app; [iTerm2](https://iterm2.com/))
- Windows: Command Prompt (i.e., cmd.exe) or PowerShell (also see [Windows Terminal](https://github.com/microsoft/terminal))
- Linux: Terminal/Console (e.g., Gnome Terminal, Konsole)

Regardless of which shell you use, the command to run Python will most likely be:

`python`

In some cases (e.g., if you have both Python 2 and 3 installed), you may need to use the following command instead:

`python3`

Once you launch Python, you will be able to enter commands directly into the interpreter from within the shell and see the results interactively. However, as noted a moment ago, we will focus on working from JupyterLab instead, since it provides a more convenient way to work on data science problems with Python (particularly when it comes to things like embedding plots). To quit a running Python session and return to the shell, simply type in the `quit()` command and execute it via the enter/return key.

### Installing Jupyter

In addition to Python, we will also install Jupyter as an environment for working with Python code interactively. Similar to Python, there are multiple options for installing Jupyter (e.g., see the official [website](https://jupyter.org/install) and [installation documentation](https://jupyter.readthedocs.io/en/latest/install.html)).

Before getting into the details of installation, it's worth briefly disambiguating two different Jupyter options: JupyterLab and Jupyter Notebooks. In simple terms, JupyterLab is a modernized version of Jupyter Notebooks that will in time fully replace the latter. As such, this guide will focus on working with JupyterLab.

To install JupyterLab, the easiest approach is to use a Python package-management system such as:

- [pip](https://pypi.org/project/pip/)
- [conda](https://docs.conda.io/projects/conda/en/latest/user-guide/tasks/manage-pkgs.html)

Which system you use depends on how you installed Python. Generally speaking, pip is the de facto option as it is included natively with Python; by contrast, conda is specific to Anaconda/Miniconda distributions.

Regardless of which package manager you are using, the commands will need to be entered into the shell to run them, similar to how we discussed running Python in the previous section. To avoid a common point of confusion, be aware that conda and pip commands must be entered directly into the shell and not into an active Python session (e.g., as outlined in the [Running Python](#Running-Python) section above).

Returning to installing JupyterLab, assuming Python is already installed and set up correctly (e.g., PATH variables are set), we would simply open the shell and execute one of the following commands:

- pip: `pip install jupyterlab` (or `pip3 install jupyterlab` in certain cases)
- conda: `conda install jupyterlab`

Assuming one of these commands works and the installation completes, you are now ready to run JupyterLab. To do so, type the following command in your shell:

`jupyter lab`

This command will launch JupyterLab in your default browser, and you are now ready to move onto the next section of this unit.

#### Google Colaboratory

While ideally it's worth knowing how to install Python and Jupyter locally, there are occasionally reasons you may run into problems doing so (e.g., your computer is out-of-date, misconfigured, or just generally slow; you're on an iPad or Android tablet; etc.). - Hardware-accelerated GPU [resources](https://colab.research.google.com/notebooks/gpu.ipynb) for user with TensorFlow and other machine/deep learning models

- GitHub [integration](https://colab.research.google.com/github/googlecolab/colabtools/blob/master/notebooks/colab-github-demo.ipynb)

For an overview of the basic features, see this [guide](https://colab.research.google.com/notebooks/basic_features_overview.ipynb).

While both Google Colaboratory and Deepnote will essentially provide all of the functionality necessary to complete this bootcamp, be aware there may be minor differences in how these custom web implementations look and function compared to the official JupyterLab version (e.g., the menu layouts will be slightly different and file uploads will require additional steps for both [Colaboratory](https://colab.research.google.com/notebooks/io.ipynb) and [Deepnote](https://docs.deepnote.com/importing-and-exporing/importing-data-to-deepnote)).

### Working with Jupyter Notebooks

This document itself is a Jupyter Notebook (formally known as IPython Notebook). Similar to a Word document (e.g., with the .docx extension), a Jupyter Notebook allows you to compose and style text. Unlike a Word document however, notebooks also allow you to embed and run code, as well as see the results of code execution (e.g., text-based output, plots, etc.).

As the name implies, this bootcamp focuses on the basics of using Jupyter Notebooks to work with Python. Nevertheless, Jupyter Notebooks support a variety of languages besides Python including R, Julia, etc.

Moreover, Jupyter Notebooks (and JupyterLab) offer for a variety of advanced functionality such as support for big data (e.g., [Apache Spark](https://spark.apache.org/)), [LaTeX](https://www.latex-project.org/) integration, GitHub-style markdown, etc. For instance, we can show Einstein's field equation (EFE) using [MathJax](https://www.mathjax.org/):

$$R_{\mu \nu} - {1 \over 2} R g_{\mu \nu} + \Lambda g_{\mu \nu}= {8 \pi G \over c^4} T_{\mu \nu}$$

Again, for the sake of this bootcamp, we will focus on core functionality that will be useful for working with Python. That being said, links to additional resources will be provided from time to time.

#### The Basics

A full set of instructions for working with Jupyter Notebooks is available as part of the [official documentation](https://jupyter-notebook.readthedocs.io/en/latest/notebook.html). Compared to that guide (which is very thorough), this section will cover the basics needed to get started when working with notebooks.

To dive right in, the first thing you'll need to do once you've launched JupyterLab is to either create a new notebook or open an existing one. To create a new notebook, you can use the main menu:

File -> New -> Notebook

When you create a new notebook, you will be asked to select a kernel, which in this case should be Python 3 (likely the default). At this point, assuming all goes well, you should see a blank tab within the main JupyterLab interface on the right along the lines of "Untitled.ipynb" (where .ipynb is the file extension for a Jupyter Notebook document).

The key to working with a notebook is the creation of *cells* in which you can type. These cells come in three main flavors:

- ***Code cells***, which contain code you plan to execute
- ***Markdown cells***, which contain descriptive information and can be stylized
- ***Raw (NBConvert) cells***, which are designed for special uses such as writing LaTeX code

In terms of which cells you will be using, this guide will focus primarily on code cells, since this is where most of the core interaction with Python will take place. Other units may provide additional information regarding other cell types (e.g., writing Markdown), but technically most everything could be done exclusively using code cells (even if the results may not be as pretty).

By default, your notebook should already start with a single, empty code cell. To see how a cell works, go ahead and enter the following code (e.g., via copy & paste) in your new notebook (or just use the cell below as an example):

````python
# a basic example...
print('Hello World!')
````

Now, to actually run this cell in Python, you have a few options:

- Click on the cell to highlight it, and then in the main menu, go to Run -> Run Selected Cells
- Use a keyboard shortcut to run the cell after highlighting it:
    - Control + Enter (or Control + Return): run the current cell
    - Shift + Enter (or Shift + Return): run the current cell and advance to the next cell automatically 
- Use the play icon ("Run the selected cells and advance") in the Notebook menu below the tabs above

Running the code should display the text 'Hello World!' directly below the code cell without quotes.

Let's take a moment to discuss how this code cell works. The first line of this code is a comment (i.e., # a basic example...), which means that is not designed to be run, but rather to explain what the code below it is about. Comments start with a # symbol, although it's also possible to comment out multiple lines in Python by wrapping them using three single quotes as well (i.e., a multi-line string, which is technically data and thus should precede any code you intend to execute and display, unless you intend for the "comment" to be displayed as output in Jupyter):

```
'''
This is an example of..
multiple lines of...
comments! =)
'''
```

Good code should contain plenty of explanatory comments to give you -- as well as others who might see your code -- a clear idea of what it's doing. Including comments is especially important when working with others (e.g., in project teams) or if you need to refer back to code you wrote later (e.g., months or years down the road).

Besides these examples of comments, our code cell also includes an actual command. Specifically, the line

`print('Hello world!')`

unsurprisingly tells Python to print the supplied text as its output. Notice that the first argument of the print function is 'Hello world!' -- i.e., text encapsulated by single quotes, which is the default way of representing character string type data in Python (we'll talk more about data types in an upcoming bootcamp unit) -- despite the input data being in single quotes, it's important to recognize that the output is unquoted.

Moving back to the interface however, there are a few other points of interest worth noting. For example, you can create additional cells using the + button in the Notebook menu, as well as cut, copy, and paste them. Alternatively, you can also find these basic functions in the main menu under Edit near the top left; notice that other shortcuts from the Notebook menu appear here too, such as the option to run a cell via the aptly named Run heading.
